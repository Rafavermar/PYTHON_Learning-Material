# EXTRACTING DATA 
LOC() basado en en nombres(etiquetas), por lo que tenemos que especificar filas y columnas usando el nombre index/fial y columna

iLOC() esta basado en la posición numerica entera, por lo que tenemos que especificar filas y columnas usando el vaor de su posicion entera( posición empieza por el numero 0)

## Diferencias Loc vs iLoc

```python
Seleccionar con:
    
                  loc (nombre)      | iloc(posición)
        --------------------------------------------
    Valor   >     df.loc['zero']    | df.iloc[0]
    Lista   > df.loc[['zero','two']] | df.iloc[[0,2]
    Slicing > df.loc['zero':'two']]  | df.iloc[0:2]
    
en metodo loc > los index están incluidos en la seleccion
en metodo iloc > los index no están incluidos en la selección
                    
```

## El método sample()

```python
# extraer 10 elementos aleatorios de la columna "nationality"
df['nationality'].sample(10, random_state=99)
#randon_state=99 indica el estado de aleatoridad con un nivel 99, por lo que 
# al correr el mismo codigo varias veces, devuelve el mismo dataframe.
# si no estuviera random_state=99, el dataframe cambia cada vez que corremos el codigo
```

```python
# extraer 20% de muestra aleatoria del dataframe
df.sample(frac=0.2)
# si corremos el codigo de nuevo, obtenemos otros dataframes diferentes
```

```python
df.sample(frac=0.2, random_state=99) 
```

```python
# incrementar la muestra -> frac>1 (obs: el parametro replace tiene que ser verdadero)
# cuando fraccion es mayor a 100% tenemos obligatoriamente poner el valor replace=True
df.sample(frac=2, replace=True, random_state=33)
# genera dataframe con el doble de filas que el DF original
```

# El método apply()

este metodo se utiliza para aplicar una funcion u operacion
bien a una serie o a un dataframe.
Se aplica a todos los valores bien de una fila determinada o una serie

```python
import numpy as np
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

```python
# usar función de numpy y aplicar a la serie
df['age'].apply(np.sqrt)
```

```python
# creamos nuestra propia función y se la aplicamos a un dataframe
# vamos a calcular el indice de masa corporal imc 
# como el (peso_kg/altura_m2)
# kg/m2
def calcular_imc(row):
        return row['weight_kg'] / ((row['height_cm']/100))**2
```

```python
df.apply(calcular_imc, axis=1)
```

## Función Lambda

```python
# función básica que suma dos valores
def sum_values(a,b):
    x = a+b
    return x
```

```python
sum_values(2,3)
```

```python
# función lambda (se usa en una linea)
# se utiliza para funciones temporales, que se usan una vez
# normalmente está asociada al uso de apply
# sum_values_lambda = lambda input:output
sum_values_lambda = lambda a,b:a+b
```

```python
sum_values_lambda(2,3)
```

## Apply + Funcion lambda

la ventaja de usar apply + lambda vs hacer referencia solo a un atributo
es que  usando .apply(lambda... podemos aplicar directamente el metodo,
sin tener que recordar el atributo.
Pues pandas, hay que referenciar el atributo y posteriormente la acción a realizar. 
por tanto apply+lambda ventaja 

### usar función lambda para convertir serie "height_cm" a metros

```python
# otra alternativa para pasar de cm a metros
#en los siguientes ejemplos entenderemos porque debemos usar lambda
df['height_cm']/100
```

# Hacer una copia a un dataframe con el metodo copy()

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

## Metodo copy()

### deep = True

```python
# deep = True por defecto (las modificaciones a la data o indices de la copia)
df_copy = df.copy()
```

```python
# actualizar valores en dataframe original
df.loc['L. Messi', 'height_cm'] = 180
```

```python
df
```

```python
# copia vs dataframe original
df_copy
```

### deep = False

estos df se denominan shallow

```python
# deep=False (cualquier cambio a la data original sera reflejado en la copia)
df_shallow_copy = df.copy(deep=False)
```

```python
# actualizar valor dataframe original
df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
df
```

```python
df_shallow_copy
```
