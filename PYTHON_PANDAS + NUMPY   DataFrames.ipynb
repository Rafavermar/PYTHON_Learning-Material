## Agregar nueva columna con Método ASSIGN()

### Assign

```python
# Se utiliza para agregar varias columnas con una sola linea de codigo
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
#leer mismo archivo csv
df_exams = pd.read_csv ('StudentsPerformance.csv')
df_exams.head()
```

```python
# crear numeros aleatorios para nuestras nuevas columnas "Score"
score1 = np.random.randint (1, 100, size=1000)
score2 = np.random.randint (1, 100, size=1000)
```

```python
# score1
```

```python
# crear una serie usando los numeros aleatorios
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
```

```python
# series2
```

```python
# usando assign() para agregar multiples columnas
df_exams.assign(score1=series1, score2=series2)
```

```python
# Genera una copia del dataframe
# si queremos sobreescribir, entonces:
df_exams = df_exams.assign(score1=series1, score2=series2)
```

```python
# mostrar dataframe
df_exams
```

# Ordenar un DataFrame 

## Ordenar valores numericos 

```python
# ordenar por una columna
df_exams.sort_values('math score')
```

```python
#value indica orden ascendente
# false indica orden descendente
df_exams.sort_values('math score', ascending=False)
```

```python
# SORT VALUES no crea copia del dataframe automaticamente!
# orden descendente por multiples columnas
# primero ordenara por Math scores y posteriormente por reading scores
df_exams.sort_values(['math score','reading score'], ascending=False)
```

```python
# Para que se refleje el orden descendiente por multiples columnas se debe indicar inplace
```

```python
df_exams.sort_values(['math score','reading score'], ascending=False,inplace=True)
```

```python
df_exams
```

## Ordenar por valores de texto

```python
df_exams.sort_values('race/ethnicity', ascending=True,key=lambda col:col.str.lower())
```

```python
# Key = lambda
# funcion que personaliza la ordenacion.
# despues de los dos puntos: se indica la accion a relaizar.
# col.srt.lower > indica que todos los valores del tipo texto sean minusculas
```

SE ORDENAN POR TANTO LAS COLUMNAS SELECCIONADAS POR ORDEN ASCENDENTE ALFABETICO Y TODOS LOS VALORES DE TEXTO TRANSFORMADOS A MINUSCULAS PARA PODER COMPARARSE Y ORDENARSE.


# Los metodos SET_INDEX() y SORT_INDEX()

```python
import pandas as pd
```

```python
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams
```

## Crear INDEX

```python
import numpy as np
import random
```

```python
# crear valores no repetitivos para los index

new_index = np.arange(0,1000)
```

```python
# barajar o desodernar los index
random.shuffle(new_index)
```

```python
# crear una nueva columna con los nuevos index
df_exams['nuevo_index'] = new_index
df_exams
```

## set_index()

```python
# fijar columna nuevo_index como index
df_exams.set_index('nuevo_index')
# esto crea una copia del dataframe con la columna añadida
```

```python
# Para guardar los cambios en el dataframe original utilizamos inplace=true
df_exams.set_index('nuevo_index', inplace=True)
```

```python
df_exams
```

## sort_index()

```python
# ordenar dataframe por new_index
df_exams.sort_index()
```

```python
df_exams
```

# Renombrar columnas e Index con rename()

## Renombrar Columna

```python
# renombrar una columna y sobreescribir los valores del dataframe
# Crea una copia del dataframe
df_exams.rename(columns={'gender': 'Gender'})
# para sobreescribir los datos en el dataframe orginal:
df_exams=df_exams.rename(columns={'gender': 'Gender'})
# o usando inplace:
df_exams.rename(columns={'gender': 'Gender'}, inplace=True)
```

```python
# renombrar una columna y actualizar el dataframe con el argumento inplace
df_exams.rename(columns={'math score': 'MS',
                        'reading score': 'RS',
                        'writing score':'WS'}, inplace=True)
```
