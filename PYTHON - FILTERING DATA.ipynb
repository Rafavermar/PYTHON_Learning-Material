# Filtrar dataframe basado en 2 o mas condiciones
en Pandas:
and-> &
or -> |

```python
import pandas as pd
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin_1')
df_laptops.head(3)
```

```python
# encontrar laptops que sean apple y cuesten mas de 2000euros
(df_laptops['Company']=='Apple') & (df_laptops['Price_euros']>=2000)
```

## & (and)

```python
# filtrar el dataframe basado en multiples condiciones
df_laptops[(df_laptops['Company']=='Apple') & (df_laptops['Price_euros']>=2000)]
```

## | (or)

```python
# encontrar laptops Appel o Dell
df_laptops['Company']=='Apple'
df_laptops['Company']=='Dell'
(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')
```

```python
# filtrar dataframe basado en las multiples condiciones
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')]
```

```python
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')].value_counts('Company')
```

```python
# encontrar laptops Apple o Dell que cuesten mas de 2000euros
df_laptops['Company']=='Apple'
df_laptops['Company']=='Dell'
df_laptops['Price_euros']>=2000

((df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')) & (df_laptops['Price_euros']>=2000)
```

```python
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell') & (df_laptops['Price_euros']>=2000)]
```

# Crear una columna condicional con 2 o mas opciones: np.select()

```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin_1')
```

```python
df_laptops.head(3)
```

```python
import numpy as np
```

```python
# Crear un array basado en multiples niveles de precio (+2 opciones)
# crear "conditions" (condiciones) y "values" (valores)

condiciones = [
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>800) & (df_laptops['Price_euros']<=2000),
    df_laptops['Price_euros']<=800
]
valores = ['Muy caro', 'Caro', 'Barata', 'Muy barata']
# >3000, >2000, >800, <=800
```

```python
# añadir una nueva columna
df_laptops['niveles_precio'] = np.select(condiciones, valores)
```

```python
# mostrar dataframe
df_laptops
```

```python
# contar los valores en columna price_tier
df_laptops.value_counts('niveles_precio')
```

## Ejercicio

```python
# crear un array basado en multiple tamaños de pantalla ("Screen size" + 2 opciones)
# Muy grande>16, Grande>14, Pequeño>12, Muy Pequeño<12
#crear "conditions" y  "values"

condiciones = [
    df_laptops['Inches']>16,
    (df_laptops['Inches']>14) & (df_laptops['Inches']<=16),
    (df_laptops['Inches']>12) & (df_laptops['Inches']<=14),
    df_laptops['Inches']<=12
]
valores = ['Muy grande', 'Grande', 'Pequeño', 'Muy pequeño']
```

```python
# añadir nueva columna
df_laptops['Screen_size'] = np.select(condiciones, valores)
```

```python
# mostrar data frame
df_laptops
```

```python
# Contrar los valores en columna Screen_size
df_laptops.value_counts('Screen_size')
```

# Encontar filas duplicadas con el metodo .duplicated()

```python
import pandas as pd
df_laptops = pd.read_csv(('laptop_price.csv'), encoding='latin_1')
df_laptops.head(3)
```

## Ejemplo 1

```python
# encontrar duplicados en 1 columna/serie
df_laptops.duplicated('laptop_ID')
```

```python
df_laptops.duplicated('laptop_ID').value_counts()
```

```python
# mostrar elementos en dataframe con duplicados en columna 'laptop_id'
df_laptops[df_laptops.duplicated('laptop_ID')]
# podemos observar que no existen  duplicados, ya que no aparecen al imprimir dataframe y arriba al utilizar value_counts todos se encuentran en la categoria false
```

```python
# duplicados en 2 o mas columnas
df_laptops.duplicated(['Product', 'TypeName', 'Inches'])
```

```python
duplicated = df_laptops.duplicated(['Product', 'TypeName', 'Inches'])
```

```python
# muestra todos los valores duplicados (excepto por el primero - keep= 'first')
# el argumento keep de esta función indica que valor original se va a queda en la seleccion.
# por defecto el valor mantenido es el primero que aparece de los duplicado, es decir, por defecto Keep=first

df_laptops[duplicated].sort_values(['Product', 'TypeName'])
```

## Ejemplo 2. keep first, last, False
Encontrar laptops mas baratas y caras en cada compañia usando el medoto sort_values() y duplicated() (argumentos: keep"first","last", and False)

```python
# ordenar dataframes ascendente por "company" y "price"
# mas barato primero y mas caro ultimo
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
df_laptops
```

```python
# revisar todas las categorias en la columna "Company"
df_laptops.value_counts('Company')
```

### keep first

```python
# valores duplicados en la columnna "Company" (por defecto keep=First)
duplicated_first = df_laptops.duplicated('Company', keep='first')
```

```python
# mostrar dataframe con valores no duplicados en la columna "Company"
# el simbolo ~ es el simbolo para indicar lo opuesto a algo
df_laptops[~duplicated_first]
```

```python
# keep='first' (laptops mas baratas por compañia)
df_laptops[~duplicated_first][['Company', 'Price_euros']]
```

```python
# revisar todas las categorias
# df_laptops[~duplicated].value_counts('Company')
df_laptops[~duplicated_first].value_counts('Company')
```

### Keep last

```python
# keep = 'Last' , ayudará a encontrar los laptops mas caros en cada categoria
duplicated_last = df_laptops.duplicated('Company', keep='last')
# keep=False
```

```python
# keep='last'(laptops mas caras por cada compañia)
# mostrar dataframe con valores no duplicados en columna"Company"
df_laptops[~duplicated_last][['Company', 'Price_euros']]
```

### keep False

```python
# mostrando dataframe con valores no duplicados en columna "Company"
# keep=False
duplicated_false = df_laptops.duplicated('Company', keep=False)
```

```python
df_laptops[~duplicated_false]
#resultado es dataframe sin valores, 
#porque False, no mantiene ni el primer ni el ultimo duplicado, si no que todos los duplicados se eliminan
# POr tanto podemos ver que en esta columna son todos duplicados y por eso no aparecen.
```

```python
# todos los elementos en la columna "Company" se repiten al menos una vez
df_laptops.value_counts('Company')
# vemos que todas las compañias tienen almenos un duplicado
# es por esto que al definir keep=False, obtenemos Dataframe vacío
```

# Eliminar duplicados con drop_duplicates()

```python
import pandas as pd
df_laptops = pd.read_csv(('laptop_price.csv'), encoding='latin_1')
df_laptops.head(3)
```

```python
# Eliminar duplicar en 2 o mas columnas
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
#ordenar dataframe ascendente por compañia (company) y precio (price)
# mas barato primero y mas caro el ultimo
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
# mas barato: keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# mas caro: keep='last'
df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Price_euros']]
```

```python
# Argumentos
df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
# inplace: eliminar duplicados en el dataframe o devuelve copia
# con inplace nos permite sobreescribir el resultado del dataframe en lugar de obtener una copia
# ignore_index=True: el index resultado tendra etiqueta 0,1,..
```

```python
df_laptops[['Company', 'Price_euros']]
```
