```python
# cear lista de ligas
# y para concatenar cadena de texto con variables
import pandas as pd

leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

for league in leagues:
    df = pd.read_csv( root + '2223' + '/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
    
# al hacer bucle for el valor de DF se va a sobreescribir, pero no es lo que queremos.
#lo que queremos es que se guarden cada Df en cada iteración.
# para eso usamos metodo Append para guardar df en una lista vacía frames. conseguiremos guardar cada DF en la iteración.

```

# Organizar Datos en un diccionario | Método concat

```python
# crear un diccionario con nombre original de liga como llave
dict_countries = {
                'Spanish La Liga' : 'SP1', 'Spanish Segunda Division' : 'SP2', 
                'German Bundesliga' : 'D1',
                'English Premier League' : 'E0', 'English League 1' : 'E2', 'English League 2' : 'E3'
                }

```

```python
# obtener un elemento del diccionario
dict_countries['Spanish La Liga']
```

```python
#  al imprimir bucle For de un diccionario nos devuelve la LLAVE
for league in dict_countries:
    print(league)
```

```python
# al imprimir dict_countries[league], nos devuelve los valores
for league in dict_countries:
    print(dict_countries[league])
```

```python
# crear lista de ligas sobre el diccionario
leagues = ['E0', 'E2', 'E3']
frames = []
# Hacer loop a un rango de temporadas e insertar una nueva columna
# en el loop se cambió la sinstaxis a la diccionario para obtener los valores
for league in dict_countries:
    for season in range(16, 22):
        df = pd.read_csv( root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# ahora vamos a crear un nuevo diccionario donde guardar todos los dataframes
import pandas as pd
dic_historical_data = {}
# crear lista de ligas
# leagues = ['E0', 'E2', 'E3']
# si colocamos frames fuera del loop, pd.concat creará lista concatenada de todo las llaves del diccionario 
# si frames dentro del loop, solo obtendremos lista concatenada para las llaves concretas del diccionario.
# asi cada liga será independiente y la informacion de liga anterior no se refleja
# con dict_historical quedará un registro y así no se perderá la info.

for league in dict_countries:
    frames = []
    for season in range(16, 22):
        df = pd.read_csv( root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dic_historical_data[league] = df_concat
```
